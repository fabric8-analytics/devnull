```python
import json
with open('repo_names.json') as f:
    json_content = json.load(f)
    print(json.dumps(json_conent, sort_keys=True, indent=4))
```

```python
for key in json_content:
    value = json_content[key]
    value = ["https://github.com/" + v + '/' +  key for v in value]
    json_content[key] = ", ".join(value)
```

```python
print(json_content)
```

```python
with open('file_t.JSON', 'w') as f:
    f.write(json.dumps(json_content, indent=4, sort_keys=True))
```

```python
import csv
import json

with open("file_t.JSON") as infile, open('output.csv', 'w') as outfile:
    writer = csv.writer(outfile, delimiter=',')
    for line in infile:
        d = json.loads(line)
        writer.writerows(d.items())
```
