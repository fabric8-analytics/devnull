```python
import nltk
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem.porter import PorterStemmer
stop_words = set(stopwords.words('english')) 
stem = PorterStemmer()
vectorizer = TfidfVectorizer(stop_words = stop_words, use_idf=True, ngram_range=(1,3))
```

```python
neg_data = pickle.load( open( "neg_data.p", "rb" ) )
pos_data = pickle.load( open( "pos_data.p", "rb" ) )
```

```python
def preproces_doc(doc):
    filtered_sent = []
    for i in doc:
#         print(i)
        i = re.sub("[^A-Za-z0-9 ]+", '', i)
        word_lst = word_tokenize(i)
        filter_lst = [w for w in word_lst if not w in stop_words]
        stem_lst = [stem.stem(w) for w in filter_lst]
        filtered_sent.append(" ".join(stem_lst))
    return filtered_sent
```

```python
filter_neg_sent = preproces_doc(neg_data)
filter_pos_sent = preproces_doc(pos_data)
```
